### Adding a new column to a DataFrame
There are four ways to insert a new column in a DataFrame

```python
std_age=pd.DataFrame({'class 1':[22,40],'class 2':[24,50],'class 3':[20,45]})
std_age
```

### 1.With the indexing operator, [ ]

```python
std_age['class4']=[25,52]  # if class4 is already there , it will overwrite the value else it create new
std_age
```

```python
std_age['class4']=[60,34]  
std_age
```

### 2.Using the insert method

```python
std_age.insert(1,"ClassB",[25,50])
std_age
```

### 3.Using the loc indexer 

```python
std_age.loc[:,"Class5"]=[100,100] ## if Column is exist it will overwrite the data else create new column
std_age
```

### 4.Using the concat function

```python
batchxy=pd.Series([45,50])
batchxy
pd.concat([std_age,batchxy],axis=1,ignore_index=True)
```

### 3.Grouping and aggregating data: 
- Pandas provides methods for grouping and aggregating data based on certain criteria.

Aggregation is the process of summarizing a group of values into a single value.
- 1.Split the data into smaller groups that are manageable and independent of each other. This is done using the groupby method in Pandas.
- 2.Apply functions on each of these groups. We can apply any of the aggregation functions, including minimum, maximum, median, mean, sum, count, standard deviation, variance, and size.
Each of these aggregate functions calculate the aggregate value of the entire group. Note that we can also write a customized aggregation function.
- 3.Combine the results after applying functions to each group into a single combined object.

## Join method – index to index
- The join method aligns two Pandas objects based on common index values. 
- That is,it looks for matching index values in both objects and then align them vertically. 
- The default type of join for this method is a left join.
- Let us consider the following example where we join two objects.


```python
periodic_table=pd.DataFrame({'Atomic Number':[1,2,3,4,5],'Element':['Hydrogen','Helium','Lithium','Beryllium','Boron'],'Symbol':['H','He',
'Li','Be','B']})
periodic_table
```

```python
periodic_table_extended=pd.DataFrame({'Atomic Number':[8,9,10],'Element':['Oxygen','Fluorine','Neon'],'Symbol':['O','F','Ne']})
periodic_table_extended
```

```python
periodic_table.join(periodic_table_extended, lsuffix="_left",rsuffix="_reight",how='left')
```

```python
periodic_table.join(periodic_table_extended, lsuffix="_left",rsuffix="_reight",how='right')
```

```python
periodic_table.join(periodic_table_extended, lsuffix="_left",rsuffix="_reight",how='inner')
```

```python
periodic_table.join(periodic_table_extended, lsuffix="_left",rsuffix="_reight",how='outer')
```

```python
periodic_table.join(periodic_table_extended, lsuffix="_left",rsuffix="_reight",how='cross')
```
