### Adding a new column to a DataFrame
There are four ways to insert a new column in a DataFrame

```python
std_age=pd.DataFrame({'class 1':[22,40],'class 2':[24,50],'class 3':[20,45]})
std_age
```

### 1.With the indexing operator, [ ]

```python
std_age['class4']=[25,52]  # if class4 is already there , it will overwrite the value else it create new
std_age
```

```python
std_age['class4']=[60,34]  
std_age
```

### 2.Using the insert method

```python
std_age.insert(1,"ClassB",[25,50])
std_age
```

### 3.Using the loc indexer 

```python
std_age.loc[:,"Class5"]=[100,100] ## if Column is exist it will overwrite the data else create new column
std_age
```

### 4.Using the concat function

```python
batchxy=pd.Series([45,50])
batchxy
pd.concat([std_age,batchxy],axis=1,ignore_index=True)
```

## Inserting rows in a DataFrame
- There are two methods for adding rows in a DataFrame, either by using the append method or with the concat function.

```python
combined_ages=pd.DataFrame({'class 1':[22,40],'class 2':[24,50],'class 3':[20,45]})
combined_ages
```

### 1.Using the append method


```python
combined_ages=combined_ages.append({'class 1':35,'class 2':33,'class 3':21},ignore_index=True)
combined_ages
```

### 2.Using the pd.concat function

```python
combined_ages=pd.DataFrame({'class 1':[22,40],'class 2':[24,50],'class 3':[20,45]})
combined_ages
```

```python
new_row=pd.DataFrame([{'class 1':32,'class 2':37,'class 3':41}])
new_row
```

```python
new_row2=pd.DataFrame([{'class 1':52,'class 2':67,'class 3':61,'class4':89}])
new_row2
```

```python
pd.concat([combined_ages,new_row,new_row2],axis=0,ignore_index=True)
```

### Deleting a row from a DataFrame
- There are two methods for removing rows from a DataFrame – either by using a Boolean selection or by using the drop method

```python
combined_ages=pd.DataFrame({'class 1':[22,40],'class 2':[24,50],'class 3':[20,45]})
combined_ages
```

```python
combined_ages[~(combined_ages.values>40)]
```

```python
combined_ages=pd.DataFrame({'class 1':[22,40,90],'class 2':[24,50,67],'class 3':[20,45,67]})
combined_ages
```

```python
combined_ages.drop([0,2],axis=0,inplace=False)
```

```python
sales_data.query("(order_price>600) and not(Product=='iPhone')")
#Query the columns of a DataFrame with a boolean expression.
```
